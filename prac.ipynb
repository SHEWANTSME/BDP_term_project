```python
# reg_ex
import re
# ca?e -> care, cafe, cave ,,,,
p = re.compile("ca.e") # . : 하나의 문자를 의미함 -> cafe, cave는 되지만 caffe는 안됨
# ^ : 문자열의 시작을 의미 -> (^de) = desk, destination, ,,이런거
# $ : 문자열의 끝 -> (se$) > case, base, ,, 이런거

if(p.match("caffe")):
    print(p.match("caffe").group())
else:
    print("none")

def matchit(alpha):
    if alpha:
        print(alpha.group())
    else: 
        print("none")
matchit(p.match("caseless")) # case는 되니까 match된다고친다.
#print(matchit(p.match("case")))
matchit(p.search("alpha beta case")) # search는 주어진 문자열중에 일치하는게 있는지 
matchit(p.search("cavelessssese")) # search는 주어진 문자열중에 일치하는게 있는지 


def matchit2(alpha):
    if alpha:
        print(alpha.group())
        print(alpha.string)
        print(alpha.start())
        print(alpha.end())
        print(alpha.span())
    else: 
        print("none")
matchit2(p.search("alpha beta gamma"))
matchit2(p.search("cave nave save case"))
matchit2(p.search("caveless"))
matchit2(p.search("good cave"))

m = p.findall("careless") # findall : 일치하는 모든 것을 리스트 형태로 반환
print(m)




# 1. p = re.compile("원하는 형태")
# 2. m = p.match("비교할 문자열") : 주어진 문자열의 처음부터 일치하는지 확인
# 3. m = p.search("~~") : 주어진 문자열 중에 일치하는게 있는지 확인
# 4. m = p.findall("~~") : 일치하는 모든 것을 list형태로 반환

# 정규식 생긴내용
# .  (caca.eee) : cacaxeee , cacapeee , cacaleee 이런것 가능, cacappppeee 안됨
# ^  (^cacs) : 문자열의 시작,, -> cacstopia , cacsalonso, cacspique 이런거 됨
# $  (se$) : 문자열의 끝 -> case, base ㄱㄴ  ? face(x)
```
